```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
path="D:\Data Analysis\Abhinand Spotify/endsong.json"
path=path.replace("\\","/")
path
```

```python
df=pd.read_json(path)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
user_id=df["username"].iloc[0]
user_id
```

```python
#converting ts to date time
df["ts"]=df["ts"].astype("datetime64")
df.head(1)
```

```python
#cheching the unique value in each columns
col=['conn_country','episode_name','episode_show_name','incognito_mode',
     "spotify_episode_uri","spotify_track_uri","user_agent_decrypted",
    "ip_addr_decrypted","skipped"]
for i in col:
    print(i,"\n",[df[i].value_counts()])
```

```python
#dropping the above columns as they are not needed for the analysis
df.drop(col,axis=1,inplace=True)
df.columns
```

```python
#renaming the columns
new_col_name={"ts":"Date_Time",
              "master_metadata_track_name":"track_name",
              "master_metadata_album_artist_name":"artist_name",
              "master_metadata_album_album_name":"album_name"
              }
df.rename(columns=new_col_name,inplace=True)
```

```python
#adding a new columns for the time played in seconds
df["s_time"]=df["ms_played"]//1000
```

```python
df.head()
```

```python
#adding a seperat column for date
df["Date"]=df["Date_Time"].dt.date
```

```python
#adding a new column for time
df["Time"]=df["Date_Time"].dt.time
```

```python
#adding a new column for Each hour the day (00:00-5:59,06:00-11:59,.....)
```

```python
df.head()
```

```python
df["reason_start"].value_counts()
```

```python
df["platform"].value_counts()
```

```python
#correting the platforsm to windows app
re_win_app=["Windows 10 (10.0.22621; x64; AppX)","Windows 10 (10.0.22000; x64)","Windows 10 (10.0.22000; x64; AppX)"]
df["platform"].replace(re_win_app,"windows app",inplace=True)
```

```python
#correting the platforsm to Android app
re_and_app=["Android OS 12 API 31 (OnePlus, GM1901)","Android OS 11 API 30 (OnePlus, GM1901)"]
df["platform"].replace(re_and_app,"Android app",inplace=True)
```

```python
re_web_app=["web_player windows 10;chrome 107.0.0.0;desktop","web_player windows 10;chrome 105.0.0.0;desktop"]
df["platform"].replace(re_web_app,"Web player",inplace=True)
```

```python
df["platform"].value_counts()
```

```python
#adding a new coulm for the time of hr

df["Hour"]=df["Date_Time"].dt.hour
```

```python

df.head()
```

```python
#adding a colum for the day of the week and month
df["week_day"]=df["Date_Time"].dt.dayofweek
df["Month"]=df["Date_Time"].dt.month
```

```python
save_path="D:/Data Analysis/Abhinand Spotify/Abhinands_cleaned_user_data.csv"
df.to_csv(save_path)
```
